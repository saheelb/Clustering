```python
#data slicing 
x = data.iloc[:, [3, 4]].values
#data shape
x.shape
```

THE ELBOW METHOD IS A HUERISTIC METHOD USED FOR DETERMINING THE CLUSTERS TO BE CONSIDERED DURING CLUSTERING ALGORITHMS SPECIALLY FOR kMeans 



```python
#empty list 
whiskers = []

#check for every tickets from 1 to 11
for tickets in range(1, 11):
    #check KMeans Clustering 
    kmns = KMeans(n_clusters = tickets, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmns.fit(x)
    whiskers.append(kmns.inertia_)

plt.plot(range(1, 11),whiskers)
plt.title('The Famous Elbow Method')
plt.xlabel('Number of Wobles(clusters)')
plt.ylabel('whiskers')
plt.show()
```

KMEANS CLUSTERING

CLUSTERING ON Annual Income(k$)

```python
kmns = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
ym = kmns.fit_predict(x)

plt.scatter(x[ym == 0, 0], x[ym == 0, 1], s = 50, c = '#ffd58e', label = 'Cluster-I')
plt.scatter(x[ym == 1, 0], x[ym == 1, 1], s = 50, c = '#ff9b5a', label = 'Cluster-II')
plt.scatter(x[ym == 2, 0], x[ym == 2, 1], s = 50, c = '#f26249', label = 'Cluster-III')
plt.scatter(x[ym == 3, 0], x[ym == 3, 1], s = 50, c = '#ec2f3b', label = 'Cluster-IV')
plt.scatter(x[ym == 4, 0], x[ym == 4, 1], s = 50, c = '#beefff', label = 'Cluster-V')
plt.scatter(kmns.cluster_centers_[:,0], kmns.cluster_centers_[:, 1], s = 80, c = 'black' , label = 'Centroid')

plt.title('K Means Clustering on Annual Income (k$)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
